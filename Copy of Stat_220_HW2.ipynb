Part d. Assume that the pitcher is randomly chosen from the team where 80% of the pitchers are right handed. Use Bayes rule to determine the probability the pitcher is right handed given that the batter hits 3 out of 4 pitches.

```python

p_rh = 0.8
p_lh = 0.2

# Probability of 3 out of 4 hits against right-handed pitcher
p_3_out_of_4_rh = binom.pmf(3, 4, 0.24)

# Probability of 3 out of 4 hits against left-handed pitcher
p_3_out_of_4_lh = binom.pmf(3, 4, 0.375)

# Bayes' Theorem
p_rh_given_3_out_of_4 = (p_3_out_of_4_rh * p_rh) / ((p_3_out_of_4_rh * p_rh) + (p_3_out_of_4_lh * p_lh))

print(f"The probability the pitcher is right-handed given that the batter hits 3 out of 4 pitches is: {p_rh_given_3_out_of_4:.4f}")
```
